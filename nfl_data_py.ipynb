## Dash

```python
app = Dash(__name__)

app.layout = html.Div([
    html.H4('Analyzing Snap Counts for Flex positions (RB, WR, TE)'),
    dcc.Dropdown(
        id="team dropdown",
        options = weekly_2022_2023['recent_team'].unique(),
        value="ARI",
        clearable=False,
    ),
    dcc.Graph(id="graph")
])

@app.callback(
    Output("graph", "figure"),
    Input("team dropdown", "value"))
def update_bar_chart(team):
  df = weekly_2022_2023
  mask = df["recent_team"] == team
  fig = px.bar(df[mask], x="week", y="Offense Pct")

  return fig

app.run_server(debug=True)
```
